```python
# ==========================================
# RANDOM SPLIT ANIME DATASET (10K / 1K)
# ==========================================

import os
import random
import shutil

# ---- Paths ----
raw_dir = "../anime_raw"
train_dir = "./task_2_data/anime/train"
test_dir = "./task_2_data/anime/test"

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# ---- Load image files ----
all_images = [
    f for f in os.listdir(raw_dir)
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
]

print(f"Total images found: {len(all_images)}")

# ---- Set seed for reproducibility ----
random.seed(42)

# ---- Sample 11000 randomly ----
selected = random.sample(all_images, min(11000, len(all_images)))

# ---- Split ----
train_images = selected[:10000]
test_images = selected[10000:11000]

print(f"Train: {len(train_images)}, Test: {len(test_images)}")

# ---- Copy files ----
for img in train_images:
    shutil.copy(os.path.join(raw_dir, img),
                os.path.join(train_dir, img))

for img in test_images:
    shutil.copy(os.path.join(raw_dir, img),
                os.path.join(test_dir, img))

print("Dataset split complete.")
```
